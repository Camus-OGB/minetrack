```python
score = []
best_k = 1
best_score = 0
for k in range(1, 30): # Start range from 1 as best_k is initialized to 1
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X, y)
    current_score = model.score(X, y) # Get the current score
    score.append(current_score)

    # Update best_k and best_score within the loop if the current score is better
    if current_score > best_score:
        best_score = current_score
        best_k = k

# Plot the scores after the loop
plt.plot(range(1, 30), score) # Plot against the range of k values
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Score")
plt.title("KNN Score vs. Number of Neighbors")
plt.show() # Display the plot

# Print the best k and best score after the loop
print(f"Best k: {best_k}")
print(f"Best Score: {best_score}")
```
